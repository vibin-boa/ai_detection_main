```python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# üîπ Load your dataset
try:
    df = pd.read_csv("ai_human1.csv")  # Replace with your filename
except FileNotFoundError:
    raise FileNotFoundError("The file 'ai_human.csv' was not found. Please check the file path.")

# üîπ Check required columns
required_columns = {'Text', 'AI_Model'}
if not required_columns.issubset(df.columns):
    raise KeyError(f"The dataset must contain the following columns: {required_columns}")

# üîπ Label AI vs Human
ai_models = {'GPT-4', 'Claude-3', 'Gemini-1.5', 'LLaMA-3'}
df['Category'] = df['AI_Model'].apply(lambda x: 'AI' if x in ai_models else 'Human')

# üîπ Feature and Target
X = df['Text']
y_model = df['AI_Model']     # For which AI
y_binary = df['Category']    # For AI vs Human

# üîπ Vectorization
vectorizer = TfidfVectorizer(stop_words='english', max_features=3000)
X_vect = vectorizer.fit_transform(X)

# üîπ Train model to predict AI_Model (GPT, Claude, Human, etc.)
X_train, X_test, y_train, y_test = train_test_split(X_vect, y_model, test_size=0.2, random_state=42)
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# üîπ Evaluate
print("üîç Model Evaluation:")
print(classification_report(y_test, model.predict(X_test)))

# üîπ Save Model & Vectorizer
joblib.dump(model, "ai_model_classifiers.pkl")
joblib.dump(vectorizer, "vectorizers.pkl")
print("‚úÖ Model and vectorizer saved!")
```
